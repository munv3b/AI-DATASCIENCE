**Q1:** Aik `Student` class banayein jismein `name` aur `roll_number` private variables (`__`) hon. In variables ko access aur modify karne ke liye getter aur setter methods likhein.

```python
# Write your code here for Q1
class Student:
    def __init__(self, name, roll_number):
        self.__name = name
        self.__roll_number = roll_number

    def get_name(self):
        return self.__name

    def set_name(self, name):
        self.__name = name

    def get_roll_number(self):
        return self.__roll_number

    def set_roll_number(self, roll_number):
        self.__roll_number = roll_number


s = Student("Ali", 101)
print(s.get_name())
s.set_name("Ahmed")
print(s.get_name())
```

**Q2:** Aik `BankAccount` class banayein jismein `__balance` private variable ho. Aik `deposit()` aur aik `withdraw()` method banayein, check karein ke withdraw karte waqt balance zero se kam na ho.

```python
# Write your code here for Q2
class BankAccount:
    def __init__(self, balance=0):
        self.__balance = balance

    def deposit(self, amount):
        self.__balance += amount
        print("Deposited:", amount)

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            print("Withdrawn:", amount)
        else:
            print("Insufficient Balance")

    def get_balance(self):
        return self.__balance


acc = BankAccount(1000)
acc.deposit(500)
acc.withdraw(300)
print(acc.get_balance())
```

**Q3:** Aik class banayein `Employee` jismein salary private ho. Agar koi salary ko minus (negative) value set karne ki koshish kare, toh setter method 'Invalid Salary' ka error message show kare.

```python
# Write your code here for Q3
class Employee:
    def __init__(self, salary):
        self.__salary = salary

    def get_salary(self):
        return self.__salary

    def set_salary(self, salary):
        if salary < 0:
            print("Invalid Salary")
        else:
            self.__salary = salary


emp = Employee(50000)
emp.set_salary(-1000)
emp.set_salary(60000)
print(emp.get_salary())
```

**Q6:** Single Inheritance ka use karte hue aik parent class `Vehicle` (with brand attribute) aur child class `Car` (with model attribute) banayein. Car ka object bana kar dono attributes print karein.

```python
# Write your code here for Q6
class Vehicle:
    def __init__(self, brand):
        self.brand = brand


class Car(Vehicle):
    def __init__(self, brand, model):
        super().__init__(brand)
        self.model = model


c = Car("Toyota", "Corolla")
print(c.brand)
print(c.model)
```

**Q13:** Operator Overloading kya hoti hai? Do objects ko `+` operator ke zariye add karne ke liye `__add__` magic method ka use kaise karte hain? Code likhein.

```python
# Write your code here for Q13
class Number:
    def __init__(self, value):
        self.value = value

    def __add__(self, other):
        return Number(self.value + other.value)

    def __str__(self):
        return str(self.value)


n1 = Number(10)
n2 = Number(20)

result = n1 + n2
print(result)
```

**Q16:** `abc` module ka use karte hue aik abstract class `Computer` banayein jisme abstract method `process()` ho. Is class ka object banane ki koshish karein aur dekhein kya hota hai.

```python
# Write your code here for Q16
from abc import ABC, abstractmethod


class Computer(ABC):

    @abstractmethod
    def process(self):
        pass


try:
    c = Computer()
except TypeError as e:
    print("Error:", e)
```

**Q18:** Aik abstract class `Car` banayein jisme aik concrete (normal) method `start_engine()` ho aur aik abstract method `drive()` ho. Child class bana kar dono ko test karein.

```python
# Write your code here for Q18
class Car(ABC):

    def start_engine(self):
        print("Engine Started")

    @abstractmethod
    def drive(self):
        pass


class Toyota(Car):
    def drive(self):
        print("Toyota is Driving")


t = Toyota()
t.start_engine()
t.drive()

```

**Q20:** Aik real-world scenario (jaise Payment System) design karein jahan `Payment` aik abstract class ho aur `CreditCardPayment` aur `UPIPayment` uski subclasses hon.

```python
# Write your code here for Q20
from abc import ABC, abstractmethod


class Payment(ABC):

    @abstractmethod
    def pay(self, amount):
        pass


class CreditCardPayment(Payment):
    def pay(self, amount):
        print(f"Paid {amount} using Credit Card")


class UPIPayment(Payment):
    def pay(self, amount):
        print(f"Paid {amount} using UPI")


cc = CreditCardPayment()
upi = UPIPayment()

cc.pay(5000)
upi.pay(2500)
```
