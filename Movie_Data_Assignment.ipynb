### **Question 4**
Which movies were released after the year 2000? (Year 2000 ke baad release hone wali movies filter karein).

### **Question 8**

### **Question 9**
How can we sort all movies by IMDB rating in descending order? (Movies ko highest se lowest rating ke mutabiq sort karein).

### **Question 10**

### **Question 11**
Which movie has the longest runtime? (Sabse lambi movie ka naam aur runtime minutes mein nikalye).

```python
data['Runtime'] = pd.to_numeric(data['Runtime'].str.replace(' min', ''), errors='coerce')

longest_movie = data.loc[data['Runtime'].idxmax()]

print("Movie:", longest_movie['Series_Title'])
print("Runtime:", longest_movie['Runtime'], "minutes")
```

### **Question 16**
'Gross' column mein jo missing values (NaN) hain, unko pooray column ke median value se fill (impute) karein.

```python
data['Gross'] = pd.to_numeric(data['Gross'], errors='coerce')
median_gross = data['Gross'].median()
data['Gross'] = data['Gross'].fillna(median_gross)

print("Median:", median_gross)
print(data['Gross'].isnull().sum())
```

### **Question 18**
'IMDB_Rating' aur 'Meta_score' ke darmiyan correlation coefficient maloom karein.

```python
data['IMDB_Rating'] = pd.to_numeric(data['IMDB_Rating'], errors='coerce')
data['Meta_score'] = pd.to_numeric(data['Meta_score'], errors='coerce')
print(data['IMDB_Rating'].corr(data['Meta_score']))
```

### **Question 20**

```python
data['Released_Year'] = pd.to_numeric(data['Released_Year'], errors='coerce')
movies_90_00 = data[(data['Released_Year'] >= 1990) & (data['Released_Year'] <= 2000)]

print("Total Movies:", movies_90_00.shape[0])
print(movies_90_00[['Series_Title', 'Released_Year']])
```
