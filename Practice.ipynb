```python
def get_grade(marks):
    if marks >= 90:
        return "A+"
    elif marks >= 80:
        return "A"
    elif marks >= 70:
        return "B"
    elif marks >= 60:
        return "C"
    else:
        return "Fail"

print(get_grade (95))
print(get_grade (50))
```

```python
def calculator(a, b, operator):
    if operator == "+":
        return a + b
    elif operator == "-":
        return a - b
    elif operator == "*":
        return a * b
    elif operator == "/":
        if b == 0: 
            return  "Cannot divide by zero!"
        return a / b
    
print(calculator(10, 10, "+")) 
print(calculator(10, 5, "-"))   
print(calculator(10, 5, "*"))  
print(calculator(10, 5, "/"))
```
